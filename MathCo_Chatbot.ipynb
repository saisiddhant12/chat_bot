```python
def text_processing(wiki_data):
    ###########   Converting to lower case  #######
    text_data = " ".join(x.lower() for x in wiki_data.split())
    
    ###########   Removing punctuation marks  #######
    text_data = re.sub(r'[^\w\s]','',text_data)

    ###########   Removing english stopwords  #######
    stop = stopwords.words('english')
    # Adding contexual stopwords, if any
    stop.extend(["xxxxxx"])

    text_data = " ".join(x for x in text_data.split() if x not in stop)

    #########      Stemming    ##########
    st = PorterStemmer()
    text_data = " ".join([st.stem(x) for x in text_data.split()])
    
    #########      Lemmatization    ##########
    lemma = nltk.wordnet.WordNetLemmatizer()
    text_data = " ".join(lemma.lemmatize(x) for x in text_data.split())

    #########      Sentence tokenizer    ##########
    ###########   Converting to lower case  #######
    lower_data = " ".join(x.lower() for x in wiki_data.split())
    wiki_sent_tokens = nltk.sent_tokenize(lower_data)
    wiki_word_tokens = nltk.word_tokenize(lower_data)# converts to list of words
    return text_data,wiki_sent_tokens,wiki_word_tokens

# text_data, wiki_sent_tokens,wiki_word_tokens = text_processing(wiki_data)

```
